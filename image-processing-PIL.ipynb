```python
from PIL import Image
import matplotlib.pyplot as plt
img = Image.open("/flower.jpeg")
print('mode',img.mode,'size',img.size)
colonne,ligne = img.size
imgF = Image.new(img.mode,img.size)
for i in range(ligne):
  for j in range(colonne):
    pixel = img.getpixel((j,i))
   # print("(",pixel[0],",", pixel[1],",",pixel[2],")")
   # print("pixel",pixel)
plt.imshow(img)
```

```python
img2 = img.copy()

for i in range(ligne):
    for j in range(colonne//2):
        p1 = img2.getpixel((j, i))
        p2 = img2.getpixel((colonne - j - 1, i))
        img2.putpixel((colonne - j - 1, i), p1)
        img2.putpixel((j, i), p2)

plt.imshow(img2)
```

```python
img3 = img.copy()
for i in range(ligne):
    for j in range(colonne):
        pixel = img3.getpixel((j, i))
        y=round((pixel[0] * 0.2126)+(pixel[1] * 0.7152)+(pixel[2] * 0.0722))
        pixel2 = (y,y,y)
        img3.putpixel((j, i), pixel2)

plt.imshow(img3)
plt.axis('off')
plt.show()
```

```python
img4=img3.copy()
for i in range(ligne):
    for j in range(colonne):
        pixel = img4.getpixel((j, i))
        if(pixel[0]+pixel[1]+pixel[2])>200:
          pixel = (255,255,255)
        else:
          pixel = (0,0,0)
        img4.putpixel((j, i), pixel)

plt.imshow(img4)
```

```python
img4=img.copy()
for i in range(ligne):
    for j in range(colonne):
        pixel = img.getpixel((j, i))
        r_neg = 255 - pixel[0]
        v_neg = 255 - pixel[1]
        b_neg = 255 - pixel[2]
        img4.putpixel((j, i), (r_neg, v_neg, b_neg))
plt.imshow(img4)
plt.axis('off')

plt.show()
```

```python
red_img = img.copy()
green_img = img.copy()
blue_img = img.copy()
for i in range(ligne):
    for j in range(colonne):
        pixel = img.getpixel((j, i))
        r, v, b = pixel
        red_img.putpixel((j, i), (r, 0, 0))
        green_img.putpixel((j, i), (0, v, 0))
        blue_img.putpixel((j, i), (0, 0, b))
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(blue_img)
plt.subplot(1, 3, 2)
plt.imshow(green_img)
plt.subplot(1, 3, 3)
plt.imshow(red_img)
plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open("/flower.jpeg")
colonne, ligne = img.size

new_colonne = colonne * 2
new_ligne = ligne * 2
larger_img = Image.new("RGB", (new_colonne, new_ligne))

for y in range(ligne):
    for x in range(colonne):
        pixel = img.getpixel((x, y))

        larger_img.putpixel((2 * x, 2 * y), pixel)          # (2x, 2y)
        larger_img.putpixel((2 * x + 1, 2 * y), pixel)      # (2x + 1, 2y)
        larger_img.putpixel((2 * x, 2 * y + 1), pixel)      # (2x, 2y + 1)
        larger_img.putpixel((2 * x + 1, 2 * y + 1), pixel)  # (2x + 1, 2y + 1)

plt.imshow(larger_img)
plt.show()
```
